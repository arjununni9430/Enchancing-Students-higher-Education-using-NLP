```python
import pandas as pd
df=pd.read_csv('emploment.csv')
df
```

```python
inputs = df.drop('Average Salary',axis='columns')
```

```python
target = df['Average Salary']
```

```python
inputs
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder
le_Field_of_Study = LabelEncoder()
le_Employment_Rate= LabelEncoder()
```

```python
#adding extra column in a dataset
inputs['Field_of_Study_n'] = le_Field_of_Study.fit_transform(inputs['Field of Study'])
inputs['Employment_Rate_n'] = le_Employment_Rate.fit_transform(inputs['Employment Rate (%)'])
```

```python
inputs
```

```python
inputs_n = inputs.drop(['Field of Study','Employment Rate (%)'],axis='columns')
```

```python
inputs_n
```

```python
target
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n, target)
```

```python
model.score(inputs_n,target)
```

```python
model.predict([[7,29,11]])
```
